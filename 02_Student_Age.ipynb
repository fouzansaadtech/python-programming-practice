```python
# Student Age Verification 

'''Create a function called check_age(). 

Requirements: 
Ask the user to enter their age.  
Handle ValueError if the input is not an integer.  
If the age is less than 18, print "Minor", otherwise print "Adult".  
Repeat the process using a loop until the user enters a valid age.  
Use finally to print "Age verification finished."'''

def check_age():
 while True:
    try:
     age=int(input("Enter a age:"))
     if age<18:
        print("Minor")
     else:
        print("Adult")
        break
    except ValueError:
      print("The input is not an integer")
    finally:
      print("Age verification finished")

check_age()


```
