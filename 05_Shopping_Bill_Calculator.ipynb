```python
# Shopping Bill Calculator 

#Create a function called shopping_bill(). 

'''Requirements: 

Ask the user how many items they purchased.  
Use a loop to get the price of each item.  
Handle ValueError if an invalid price is entered.  
Ask the user for the number of people to split the bill.  
Handle ZeroDivisionError if the number of people is 0.  
Use an if condition to display:  
o "High Bill" if each person's share is greater than 1000.  
o Otherwise, "Affordable Bill".  
Use finally to print "Billing process completed." '''

def shopping_bill():
    try:
        many=int(input("How many items did you purchase:"))
        items=[]

        for i in range(many):
            i=int(input(f"The price of {i+1} items:"))
            items.append(i)
        total=sum(items)

        memb=int(input("No of people split the bill:"))

        split=total/memb
        print(f"Split bill of each person:{split}")

        if split>1000:
                print("HIGH BILL")
        else:
                print("AFFORDABLE BILL")

    except ValueError:
        print("An invalid price is entered")
    except ZeroDivisionError:
        print("The number of people is 0")

    finally:
        print("Billing process completed")

shopping_bill()

```
