```python
# Number List Average 

'''Create a function called find_average(). 

Requirements: 

Ask the user how many numbers they want to enter.  
Store the numbers in a list using a loop.  
Handle:  
o ValueError for invalid number input.  
o ZeroDivisionError if the count is 0 while calculating the average.  
Use an if condition to check if the average is greater than 50.  
Use finally to print "Average calculation completed." '''

def find_average():
    try:
        number=[]
        many=int(input("How many numbers:"))

        for i in range(many):
         i=int(input(f"Enter a number:{i+1}"))
         number.append(i)

        avg=sum(number)/many

        if avg>50:
            print("Average is greater than 50")
        else:
            print("Average is less tha 50")
            
    except ValueError:
       print("Invalid number input")
    except ZeroDivisionError:
        print("The count is 0, cannot calculate average")
    finally:
       print("Average calculation completed")

find_average()
            
```
